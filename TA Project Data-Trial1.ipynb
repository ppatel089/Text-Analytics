```python
# Import the Twython class
from twython import Twython
import json
```

```python
# Load credentials from json file
with open("twitter_credentials.json", "r") as file:
    creds = json.load(file)
```

```python
# Instantiate an object
python_tweets = Twython(creds['CONSUMER_KEY'], creds['CONSUMER_SECRET'])
```

```python
# Create our query
query = {'q': 'Joe Biden',
        'result_type': 'popular',
        'count': 100,
        'lang': 'en',
        }
```

```python
import pandas as pd

# Search tweets
dict_ = {'user': [], 'date': [], 'text': [], 'favorite_count': []}
for status in python_tweets.search(**query)['statuses']:
    dict_['user'].append(status['user']['screen_name'])
    dict_['date'].append(status['created_at'])
    dict_['text'].append(status['text'])
    dict_['favorite_count'].append(status['favorite_count'])
```

```python
# Structure data in a pandas DataFrame for easier manipulation
df = pd.DataFrame(dict_)
#df.sort_values(by='favorite_count', inplace=True, ascending=False)
df
```

```python
import tweepy as tw
auth = tw.OAuthHandler(credentials['CONSUMER_KEY'], credentials['CONSUMER_SECRET'])
auth.set_access_token(credentials['ACCESS_TOKEN'], credentials['ACCESS_SECRET'])
api = tw.API(auth, wait_on_rate_limit=True)
new_search = "Joe+Biden -filter:retweets"

tweets = tw.Cursor(api.search,
                   q=new_search,
                   lang="en",
                   since='2020-01-01').items(1000)

all_tweets = [[tweet.user.screen_name,tweet.user.name,tweet.created_at,tweet.user.location,tweet.text] for tweet in tweets]
all_tweets
```

```python
tweet_text = pd.DataFrame(data=all_tweets, columns=['screen_name','user','date','location','text'])
tweet_text
```
